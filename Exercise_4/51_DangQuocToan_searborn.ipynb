```python
# Your code here
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.scatterplot(x = df['fare'], y = df['age'], data= df)
plt.title("Cau 1")
plt.show()
```

```python
sns.jointplot(x = df['fare'], y = df['age'], data= df)
plt.title("Cau 1")
plt.show()
```
